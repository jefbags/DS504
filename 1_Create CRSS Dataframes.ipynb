# ACCIDENT FILE

```python
# First, we need a new column for severe accident yes/no
# This the key column we want to model data on...

ACCIDENT.insert(18, 'SEV_ACC', 0)
ACCIDENT.loc[(ACCIDENT.MAXSEV_IM == 3) | (ACCIDENT.MAXSEV_IM == 4), "SEV_ACC"] = 1

# Test it
#ACCIDENT.loc[(ACCIDENT.MAX_SEV == 3) | (ACCIDENT.MAX_SEV == 4)]            
#ACCIDENT.loc[(ACCIDENT.SEV_ACC == 1)]
```

```python
# Now let's reduce the columns in Accident to those ones we really care about...
# Will add them to a new dataframe, will be primary for the analysis going forward

# Can also drop them...
# ACCIDENT_TEST = ACCIDENT.drop(['WEATHER'], axis=1)

# SKIPPING::::
# Using Imputed Values Instead: 'HARM_EV', 'LGT_COND', 'MAX_SEV', 'ALCOHOL', 'WEATHER', 'WEATHER1', 
#  'WEATHER2',

# Dropping from Eval.: 'HOUR', 'HOUR_IM', 'INT_HWY', 'MANCOL_IM', 'MAN_COLL', 'MINUTE', 'MINUTE_IM', 
#  'NO_INJ_IM', 'NUM_INJ', 'PEDS', 'PERMVIT', 'PERNOTMVIT',  'PJ', 'PSU', 'PSUSTRAT', 'PSU_VAR', 
#  'PVH_INVL','RELJCT1', 'RELJCT1_IM', 'RELJCT2', 'RELJCT2_IM','SCH_BUS','VE_FORMS','WEIGHT',
#  'WRK_ZONE', 'YEAR', 'file', 'VE_TOTAL',

columns_i_want=['ALCHL_IM', 'CASENUM', 'CF1', 'CF2', 'CF3',
       'EVENT1_IM', 'LGTCON_IM', 'MAXSEV_IM', 'SEV_ACC','MONTH', 
       'REGION',  'REL_ROAD','STRATUM', 'TYP_INT', 'URBANICITY', 
        'WEATHR_IM',  'WKDY_IM','WRK_ZONE']

PROJ_DF = ACCIDENT[columns_i_want]

# Can also do inline
# ACCIDENT3 = ACCIDENT3['ALCOHOL']
```

```python
# We will one-hot encode selected fields for better accuracy in the modealing.
# We will encode the following fields
# STRATUM, LGTCON_IM, EVENT1_IM, TYP_INT, REL_ROAD, WRK_ZONE, WEATHER_IM, ALCOHOL_IM, REGION, URBANICITY

# AND, we need to do an extra step with these becasue it's the same field three times 
# (accounts for mult factors). CF1, CF2 and CF3

# So, we need to one hot encode, then sum and combone into singular columns
```

```python
# CF1, CF2 and CF3

# Let's start with the special case, the Related Factors- Crash Level
# The goal will be to one hot encode, sum and combine into a new column

PROJ_DF = pd.get_dummies(PROJ_DF, columns=['CF1', 'CF2', 'CF3'])
```

```python
# Now we need to find max in the duplicates, and combine into new column
#EXAMPLE: sum_column = df["col1"] + df["col2"]
#EXMAPLE: df["col3"] = sum_column

# MAKING A DECSION NOT TO INCLUDE UNKNOWNS!!
# SKIPPING 0 and 99 for 'None' and 'Uknown'!
# COME back to this - it might not be the best decision

nums = [3,5,7,12,13,14,15,16,17,19,20,21,23,24,25,26,27,28]

for i in nums:
    #print (i)
    this_CF = PROJ_DF.filter(regex="_" +str(i))
    PROJ_DF['CF_'+str(i)] = this_CF.max(axis=1)

# And then delete all the old columns 
PROJ_DF = PROJ_DF.loc[:, ~PROJ_DF.columns.str.startswith('CF1_')]
PROJ_DF = PROJ_DF.loc[:, ~PROJ_DF.columns.str.startswith('CF2_')]
PROJ_DF = PROJ_DF.loc[:, ~PROJ_DF.columns.str.startswith('CF3_')]

#https://stackoverflow.com/questions/12569730/sum-all-columns-with-a-wildcard-name-search-using-python-pandas
```

```python
# Now lets one hot encode the others
# STRATUM, LGTCON_IM, EVENT1_IM, TYP_INT, REL_ROAD, WRK_ZONE, WEATHER_IM, ALCOHOL_IM, REGION, URBANICITY

PROJ_DF = pd.get_dummies(PROJ_DF, columns=['STRATUM','LGTCON_IM','EVENT1_IM','TYP_INT','REL_ROAD','WRK_ZONE','WEATHR_IM','ALCHL_IM','REGION','URBANICITY', 'MONTH', 'WKDY_IM'])
```

```python
# Drop columns with 99 (Uknown, not reported
# I AM GOING TO DROP ALL THE 'UKNOWN' CODED COLUMNS....
# Why?  Because who cares if something is uknown - it shouldn't be a factor in assessing the crash
PROJ_DF = PROJ_DF.drop(['TYP_INT_98','TYP_INT_99','REL_ROAD_98','REL_ROAD_99'], axis=1)
```

```python
# Delete every columns with all zeroes.... why keep? Not factors in the crashes...

# https://stackoverflow.com/questions/21164910/how-do-i-delete-a-column-that-contains-only-zeros-in-pandas
# df = df.loc[:, (df != 0).any(axis=0)]

# Are there any? No... There are not
PROJ_DF.loc[:, (PROJ_DF == 0).all()]
```

# VEHICLE FILE

```python
# Now we need to start working on the ancillary files.
# We have : VEHICLE, PERSON, DRIMPAIR, DISTRCT, FACTOR
# Let's start with VEHICLE, 

# First, I want to see if older cars are more prone to severe accidents
# Let's create a new column and assign cars older than 2010 to it

VEHICLE.insert(88,'OLD_CAR', 0)
VEHICLE.loc[(VEHICLE.MOD_YEAR < 2011), "OLD_CAR"] = 1

# Then lets add new columns for speed
VEHICLE.insert(89,'SPD_L30MPH', 0)
VEHICLE.insert(90,'SPD_30_65MPH', 0)
VEHICLE.insert(91,'SPD_G65MPH', 0)

VEHICLE.loc[(VEHICLE.TRAV_SP < 30), "SPD_L30MPH"] = 1
VEHICLE.loc[(VEHICLE.TRAV_SP >= 30) & (VEHICLE.TRAV_SP <= 65), "SPD_30_65MPH"] = 1
VEHICLE.loc[(VEHICLE.TRAV_SP > 65), "SPD_G65MPH"] = 1
```

```python
# Here are the fields we are going to work on
# I'm skipping ones too specific on vehicle make/model etc. Want to focus on crash causation 


columns_i_want=['BDYTYP_IM', 'CASENUM', 'OLD_CAR', 'SPD_L30MPH', 'SPD_30_65MPH', 'SPD_G65MPH', 'SPEEDREL', 'DR_SF1','DR_SF2',
                'DR_SF3','DR_SF4','VALIGN','VPROFILE','VSURCOND','PCRASH1_IM']

VEHICLE= VEHICLE[columns_i_want]
```

```python
# Now lets one hot encode the fields

VEHICLE = pd.get_dummies(VEHICLE, columns=['BDYTYP_IM', 'DR_SF1', 'DR_SF2', 'DR_SF3', 'DR_SF4', 'SPEEDREL','VALIGN','VPROFILE','VSURCOND','PCRASH1_IM'])
```

```python
# As with accident, there are three fields that contain the same data, we need to combine and take max

# SKIPPING 0 and 99 for 'None' and 'Unknown'!
nums = [6, 8, 9, 10, 16, 18, 20, 21, 22, 23, 24, 32, 36, 37, 50, 51, 54, 55, 56, 57, 58, 59, 60, 91]

for i in nums:
    #print (i)
    this_DR = VEHICLE.filter(regex="_" +str(i))
    VEHICLE['DR_SF_'+str(i)] = this_DR.max(axis=1)

# And then delete all the old columns 
VEHICLE = VEHICLE.loc[:, ~VEHICLE.columns.str.startswith('DR_SF1_')]
VEHICLE = VEHICLE.loc[:, ~VEHICLE.columns.str.startswith('DR_SF2_')]
VEHICLE = VEHICLE.loc[:, ~VEHICLE.columns.str.startswith('DR_SF3_')]
VEHICLE = VEHICLE.loc[:, ~VEHICLE.columns.str.startswith('DR_SF4_')]
```
