```python
# The files all need to be in the same directory.  So, start by renaming them so there's no conflict
# Also: The FARS has mixed case data.... it's reeking havoc.  Start by making everything upper case

# Do for all dirs/years

year = 2010
for i in range(9):
    dir = '/home/home/Desktop/DS504/data/FARS/FARS'+str(year)+'NationalCSV/'
    for count, filename in enumerate(os.listdir(dir)):
        #print (count)
        #print(filename)
        dst = str(year)+"_" + filename
        dst = dst.upper()  
        src = dir + filename
        dst = dir + dst
        os.rename(src, dst)
    year = year + 1

print("success")
```

```python
# Now we need to move or copy them all to the combined folder of the same name...

year2 = 2010
for j in range(9):
    dir = '/home/home/Desktop/DS504/data/FARS/FARS'+str(year2)+'NationalCSV/'
    os.chdir(dir)
    for file in glob.glob("*.*"):
        print(file)
        dstdir = '/home/home/Desktop/DS504/data/FARS/COMBINEDFARS/'
        dstfolder = file.replace(str(year2)+"_","").replace(".CSV","")
        dst = dstdir+dstfolder
        shutil.copy(file,dst)
    year2 = year2 + 1
```
